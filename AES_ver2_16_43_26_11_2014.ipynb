```python
def str_to_hex_arr(in_str):
    out_hex = np.empty([4,4],dtype='S2')
    index = 0
    for c in range(4):
        for r in range(4):
            out_hex[r][c] = in_str[index*2:index*2+2]
            index = index+1
    return out_hex

def hex_to_int(in_str):
    o_int = np.zeros((in_str.shape[0],in_str.shape[1]))
    for r in range(in_str.shape[0]):
        for c in range(in_str.shape[1]):
            o_int[r][c] = int(in_str[r][c],16)
    o_int = o_int.astype(int)
    return o_int

def int_to_hex(input_int):
    vhex = np.vectorize(hex)
    return vhex(input_int)
```
